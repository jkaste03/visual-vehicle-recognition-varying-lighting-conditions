```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("John-Oscar.csv")
df = df[["color", "image",
         "lighting", "model", "year_3", "year_s", "year_x", "year_y"]]

y = df[df["model"] == "Y"]
x = df[df["model"] == "X"]
e = df[df["model"] == "3"]
s = df[df["model"] == "S"]
```

```python

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.countplot(x='year_y', data=y, ax=axes[0, 0], hue="year_y")
sns.countplot(x='year_x', data=x, ax=axes[0, 1], hue="year_x")
sns.countplot(x='year_3', data=e, ax=axes[1, 0], hue="year_3")
sns.countplot(x='year_s', data=s, ax=axes[1, 1], hue="year_s")

plt.tight_layout()
plt.show()
```
