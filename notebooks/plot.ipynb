```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../annotations/internal.csv')

y = df[df['model'] == 'Y']
x = df[df['model'] == 'X']
e = df[df['model'] == '3']
s = df[df['model'] == 'S']

other = 'Other car'
```

```python
data={'Tesla': len(df[df['model'] != other]), 'Other': len(df[df['model'] == other])}
plot = sns.barplot(data=data)
plot.set_title('Fordeling av Tesla og andre biler')
plt.show()
```

```python
data = df[df['model'] != other]
plot = sns.countplot(data=data, x='model', hue='model')
plot.set_title('Modellfordeling av Teslaer')
plt.show()
```

```python
data = df[df['model'] != 'other']
crosstab = pd.crosstab(data['model'], data['year'])
crosstab.plot(kind='bar', stacked=True)

plt.title('Modellfordeling av Teslaer')
plt.xlabel('Modell')
plt.ylabel('Count')
plt.show()
```

```python

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Årsfordeling')

sns.countplot(x='year', data=y, ax=axes[0, 0], hue='year')
sns.countplot(x='year', data=x, ax=axes[0, 1], hue= 'year')
sns.countplot(x='year', data=e, ax=axes[1, 0], hue='year')
sns.countplot(x='year', data=s, ax=axes[1, 1], hue='year')

axes[0,0].title.set_text('Model Y')
axes[0,1].title.set_text('Model X')
axes[1,0].title.set_text('Model 3')
axes[1,1].title.set_text('Model S')

plt.tight_layout()
plt.show()
```

```python
fig = sns.countplot(data=df[df['model'] == other], x='color', hue='color')
```
